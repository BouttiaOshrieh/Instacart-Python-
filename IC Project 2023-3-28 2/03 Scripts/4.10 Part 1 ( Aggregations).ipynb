#### 2. Customer profile aggregations 

```python
#age and family status aggregations
age_family_aggs = all_merged.groupby('age_family_profile').agg({'prices': ['mean', 'min', 'max'], 'days_since_prior_order': ['mean', 'min', 'max']})
```

```python
#Average product price by age and family status
age_family_mean = all_merged.groupby('age_family_profile').agg({'prices': 'mean'})
print(age_family_mean)
```

```python
#most sold product among different groups of age and family status
age_fam_prods_mode = all_merged.groupby('age_family_profile')['product_name'].apply(lambda x: x.mode()[0])
nunique_mode_values = all_merged.groupby('age_family_profile')['product_name'].apply(lambda x: x.mode().nunique())
print(age_fam_prods_mode)
print(nunique_mode_values)
```

```python
#total sales by age and family status
sales_age_fam = all_merged.groupby('age_family_profile')['prices'].sum()
print(sales_age_fam)
```
